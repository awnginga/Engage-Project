```python
import pandas as pd # Import pandas if you haven't already

# Assuming your original DataFrame is in a CSV file, for example
df = pd.read_csv('/content/nhanes_reduced .csv')

#Group by HHIncome and Education
Grouped_data = df.groupby(['HHIncome','Education']).size()
print(Grouped_data)
```

```python
#Group by HHIncome and Education
Grouped_data = df.groupby(['HHIncome','Education']).size().unstack()
print(Grouped_data)
Grouped_data.plot(kind='bar', figsize=(10,6))
plt.xlabel('HHIncome')
#plt.xlabel('Count')
plt.xticks(rotation=45)
plt.ylabel('Education')
plt.tight_layout()

plt.show()
```

```python
#stacked bar chart
Grouped_data.plot(kind='bar', stacked=True, figsize=(10,6))
plt.xlabel('HHIncome')
#plt.xlabel('Count')
plt.xticks(rotation=45)
plt.ylabel('Education')
plt.tight_layout()
```

```python
# Get unique age grroup
Education_groups=df['Education'].unique()
print(Education_groups)
#Create subplot
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18,10),sharey=True)
axes=axes.flatten()
```

```python
#scatter plot for age and AgeMonths
plt.scatter(df['Age'],df['AgeMonths'], color='Grey')
plt.xlabel('Age')
plt.ylabel('AgeMonths')
plt.title('Age vs AgeMonths')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing 'Age' and 'MaritalStatus' columns

# Create a histogram for age
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()

# Create a bar chart for marital status
marital_status_counts = df['MaritalStatus'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(marital_status_counts.index, marital_status_counts.values)
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Distribution of Marital Status')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

```
