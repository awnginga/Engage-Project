```python
from typing import final
#reading the data
final=pd.read_excel("/content/drive/MyDrive/ENGAGE/Datasets/MultiLinearRegressionPractical/child_data200.xlsx")
#displaying the first five rows
final.head()
```

```python
#Creating the x values
x=final.drop(columns="Child's length or height (centimetres)")
```

```python
#creating the output variable
y=final["Child's length or height (centimetres)"]
```

```python
#Creating the model
model=linear_model.LinearRegression()
```

```python
#spliting the data set into train and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#training the model using the training data data
model.fit(x_train,y_train)
```

```python
#predicting the values for the x_testing values
y_predict=model.predict(x_test)
```

```python
#Evaluating the performance of the model
from sklearn.metrics import r2_score, mean_absolute_error
print("R_squared")
r_square=r2_score(y_test,y_predict)
adjusted_r2 = 1 - (1 - r_square) * (402 - 1) / (402 - 5 - 1)
print("Adjusted R", adjusted_r2)
print(r2_score(y_test,y_predict))
print("Mean absolute error")
print(mean_absolute_error(y_test,y_predict))
```
